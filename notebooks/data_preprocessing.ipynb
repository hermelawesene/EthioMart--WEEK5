```python
import pandas as pd
import re

# Load your DataFrame
df = pd.read_csv('NejashMarket.csv')
```

```python
df
```

```python
print("Checking for NaN values in the 'Message' column:")
nan_count = df['message_text'].isnull().sum()
print(f"Number of NaN values in 'Message' column: {nan_count}")
```

```python
df = df.dropna(subset=['message_text'])

# Print the shape of the dataset after dropping NaN values in the "Message" column
print(f"Dataset shape after dropping NaN values in 'Message' column: {df.shape}")
```

```python
# Define a function to remove emojis
def remove_emojis(text):
    # Check if the input is a string before processing
    if isinstance(text, str):
        emoji_pattern = re.compile(
            "[" 
            "\U0001F600-\U0001F64F"  # emoticons
            "\U0001F300-\U0001F5FF"  # symbols & pictographs
            "\U0001F680-\U0001F6FF"  # transport & map symbols
            "\U0001F700-\U0001F77F"  # alchemical symbols
            "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
            "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
            "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
            "\U0001FA00-\U0001FA6F"  # Chess Symbols
            "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
            "\U00002702-\U000027B0"  # Dingbats
            "\U000024C2-\U0001F251" 
            "]+", 
            flags=re.UNICODE
        )
        return emoji_pattern.sub(r'', text)
    else:
        return text  # Return non-string values as is (could also use '' if you want empty strings)

# Apply the function to the 'message_text' column
df['message_text'] = df['message_text'].apply(remove_emojis)

# Display the updated DataFrame
print(df.head())
```
